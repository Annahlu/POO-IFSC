# Método TOP

```python
def string_to_dict(S):
    S = np.array(list(S))
    d = {}
    for s in np.unique(S):
        d[s] = np.sum(S == s)
    return d

def isanagrama(p1, p2):
    return string_to_dict(p1) == string_to_dict(p2)
        
def anagrama(P):
    sizes = np.array(list(map(lambda x : len(x), P)))
    ascii_sums = np.array(list(map(lambda x : np.sum(list(map(ord,list(x)))), P)))
    ascii_prods = np.array(list(map(lambda x : np.prod(list(map(ord,list(x)))), P)))
    passed = np.zeros(len(P))
    i = 0
    palavras = []
    for p1 in P:
        if(passed[i] == 0):
            anagramas = [p1]
            passed[i] = 1
            pos = np.where((sizes == sizes[i]) & (passed == 0) & (ascii_sums == ascii_sums[i]) & (ascii_prods == ascii_prods[i]))[0]
            for j in pos:
                if(isanagrama(p1, P[j])):
                    anagramas.append(P[j])
                    passed[j] = 1
            palavras.append(anagramas)
        i += 1
    return palavras
```

```python
f1 = open('test1', encoding='utf-8')
f2 = open('test2', encoding='utf-8')
f3 = open('test3', encoding='utf-8')

x1 = list(map(lambda x : x.replace('\n','').lower(), f1.readlines()))
x2 = list(map(lambda x : x.replace('\n','').lower(), f2.readlines()))
x3 = np.array(list(map(lambda x : x.replace('\n','').lower(), f3.readlines())))
```

```python
start_time = time.time()
a1 = anagrama(x1)

print("Elapsed time:", time.time()-start_time)

a1
```

```python
start_time = time.time()
a2 = anagrama(x2)

print("Elapsed time:", time.time()-start_time)

a2
```

```python
start_time = time.time()
a3 = anagrama(x3)

print("Elapsed time:", time.time()-start_time)

a3
```

```python
def string_to_number(S):
    dic = {'a':2, 'b':3, 'c':5, 'ç':7, 'd':11, 'e':13, 'f':17, 'g':19, 'h':23, 'i':29, 'j':31, 'k':37, 'l':41, 'm':43, 'n':47, 'o':53, 'p':59, 'q':61, 'r':67, 's':71, 't':73, 'u':79, 'v':83, 'w':89, 'x':97, 'y':101, 'z':103, 'á':107, 'â':109, 'ã':113, 'ó':127, 'õ':131, 'ô':137, 'é':139, 'ê':149, 'è':151, 'í':157, 'ú':163, 'û':167,'\'':173,'ñ':179,'ä':181,'ö':191,'å':193}
    return list(map(lambda x : np.prod(list(map(lambda y : dic[y], list(x)))), S))
```

```python
def anagramaTOP(P):
    P = np.array(P)
    N = np.array(string_to_number(P))
    passed = np.zeros(len(P))
    
    i = 0
    palavras = []
    for i in range(len(P)):
        if(passed[i] == 0):
            pos = np.where(N == N[i])
            anagramas = list(P[pos])
            passed[i] = 1
            passed[pos] = 1
            palavras.append(anagramas)
            
    return palavras
```
